```python
class Node:
    def __init__(self, data):
        self.data = data;
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head is None
    
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def insert_after_node(self, prev_node, data):
        """Insert a new node after a specific node."""
        if not prev_node:
            print("Previous node cannot be None")
            return
        
        new_node = Node(data)
        new_node.next = prev_node.next
        prev_node.next = new_node
    
    def insert_at_position(self, position, data):
        if position < 0:
            print("Position should be non-negative")
            return
        if position == 0:
            self.insert_at_beginning(data)
            return
    
        new_node = Node(data)
        current = self.head
        pos = 0
        while current and pos< position-1:
            current= current.next
            pos+=1
        
        if not current:
            print(f"Position {position} out of bounds")
            return

        new_node.next = current.next
        current.next = new_node

    def delete_at_beginning(self):
        if self.is_empty():
            print("List is empty")
            return None
        temp = self.head 
        self.head = self.head.next
        deleted_data = temp.data
        temp =None
        return deleted_data

    def delete_at_end(self):
        if self.is_empty():
            print("Empty")
            return None
    
        if not self.head.next:
            temp = self.head
            self.head = None
            return temp.data
    
        current = self.head
        while current.next.next:
            current = current.next
        deleted_data = current.next.data
        current.next = None
        return deleted_data

    def delete_node(self, key):
        """Delete a node with a specific value."""
        if self.is_empty():
            print("List is empty")
            return
        
        # If the head node itself holds the key to be deleted
        if self.head.data == key:
            self.head = self.head.next
            return
        
        # Search for the key to be deleted, keeping track of the previous node
        current = self.head
        while current.next and current.next.data != key:
            current = current.next
        
        # If key was not present in linked list
        if not current.next:
            print(f"Key {key} not found")
            return
        
        # Unlink the node from linked list
        current.next = current.next.next

    
    def delete_at_position(self, position):
        """Delete a node at a specific position."""
        if self.is_empty():
            print("List is empty")
            return None
        
        if position < 0:
            print("Position should be non-negative")
            return None
        
        # If the position is 0, delete the head
        if position == 0:
            return self.delete_at_beginning()
        
        current = self.head
        pos = 0
        
        # Traverse to the position
        while current and pos < position - 1:
            current = current.next
            pos += 1
        
        # If position is out of bounds
        if not current or not current.next:
            print(f"Position {position} out of bounds")
            return None
        
        # Delete the node at position
        deleted_node = current.next
        deleted_data = deleted_node.data
        current.next = deleted_node.next
        deleted_node = None
        return deleted_data
    
    def get_at_position(self, position):
        """Access a node at a specific position."""
        if self.is_empty():
            print("List is empty")
            return None
        
        if position < 0:
            print("Position should be non-negative")
            return None
        
        current = self.head
        pos = 0
        
        # Traverse to the position
        while current and pos < position:
            current = current.next
            pos += 1
        
        # If position is out of bounds
        if not current:
            print(f"Position {position} out of bounds")
            return None
        
        return current.data
    
    def search(self, key):
        """Search for a node with a specific value and return its position."""
        if self.is_empty():
            print("List is empty")
            return -1
        
        current = self.head
        position = 0
        
        # Traverse the list
        while current:
            if current.data == key:
                return position
            current = current.next
            position += 1
        
        return -1  # Not found
    def display(self):
        """Display the contents of the linked list."""
        if self.is_empty():
            print("List is empty")
            return
        
        current = self.head
        while current:
            print(current.data, end=" -> " if current.next else "\n")
            current = current.next
    
    def length(self):
        """Return the length of the linked list."""
        count = 0
        current = self.head
        
        while current:
            count += 1
            current = current.next
        
        return count
    
    def reverse(self):
        """Reverse the linked list in-place."""
        prev = None
        current = self.head
        
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        
        self.head = prev
```
