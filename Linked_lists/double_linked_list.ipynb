```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None # Optional we maintain a tail pointer for O(1) operations at end
    
    def is_empty(self):
        return self.head is None
    
    def insert_at_beginning(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
            self.tail = new_node
            return 
        new_node_next = self.head
        self.head.prev = new_node
        self.head = new_node
    
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = new_node
            self.tail = new_node
            return
        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node

    def insert_after_node(self, prev_node, data):
        if not prev_node:
            print("Previous node cannot be None")
            return
        new_node = Node(data)
        if prev_node == self.tail:
            self.insert_at_end(data)
            return
        new_node.next = prev_node.next
        new_node.prev = prev_node
        prev_node.next.prev = new_node
        prev_node.next = new_node

    def insert_before_node(self,next_node, data):
        new_node = Node(data)
        if not next_node:
            print("Next node cannot be None")
            return
        new_node.prev = next_node.prev
        new_node.next = next_node
        next_node.prev.next = new_node
        next_node.prev = new_node
    
    def insert_at_position(self, position, data):
        if position < 0:            
            print("Position should be non-negative")
            return

    
        if position == 0:
            self.insert_at_beginning(data)
            return
        
        current = self.head
        pos = 0
        
        # Traverse to the position
        while current and pos < position:
            current = current.next
            pos += 1
        
        # If position is beyond the end, insert at the end
        if not current:
            self.insert_at_end(data)
            return
        
        # Insert before the current node
        self.insert_before_node(current, data)
    

    def delete_at_beginning(self):
        deleted_data = self.head.data

        if self.head == self.tail:
            self.head = None
            self.tail = None
            return deleted_data

        self.head = self.head.next
        self.head.prev = None
        return deleted_data
    
    def delete_at_end(self):
        deleted_data = self.tail.data
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return deleted_data
        
        self.tail = self.tail.prev
        self.tail.next = None
        return deleted_data
    
    def delete_node(self, node):
        """Delete a specific node."""
        if not node:
            print("Node cannot be None")
            return
        
        # If it's the only node
        if self.head == self.tail == node:
            self.head = None
            self.tail = None
            return
        
        # If it's the head
        if node == self.head:
            return self.delete_at_beginning()
        
        # If it's the tail
        if node == self.tail:
            return self.delete_at_end()
        
        # Adjust pointers to skip over the node
        node.prev.next = node.next
        node.next.prev = node.prev
        return node.data
    


    def delete_by_value(self, key):
        current = self.head
        while current and current.data!=key:
            current = current.next
        self.delete_node(current)
        
        def delete_at_position(self, position):
        """Delete a node at a specific position."""
        if self.is_empty():
            print("List is empty")
            return None
        
        if position < 0:
            print("Position should be non-negative")
            return None
        
        # If the position is 0, delete the head
        if position == 0:
            return self.delete_at_beginning()
        
        current = self.head
        pos = 0
        
        # Traverse to the position
        while current and pos < position:
            current = current.next
            pos += 1
        
        # If position is out of bounds
        if not current:
            print(f"Position {position} out of bounds")
            return None
        
        return self.delete_node(current)
        def get_at_position(self, position):
        """Access a node at a specific position."""
        if self.is_empty():
            print("List is empty")
            return None
        
        if position < 0:
            print("Position should be non-negative")
            return None
        
        current = self.head
        pos = 0
        
        # Traverse to the position
        while current and pos < position:
            current = current.next
            pos += 1
        
        # If position is out of bounds
        if not current:
            print(f"Position {position} out of bounds")
            return None
        
        return current.data
    
      def search(self, key):
        """Search for a node with a specific value and return its position."""
        if self.is_empty():
            print("List is empty")
            return -1
        
        current = self.head
        position = 0
        
        # Traverse the list
        while current:
            if current.data == key:
                return position
            current = current.next
            position += 1
        
        return -1  # Not found
    
```
